```python
df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce'
)
```

```python
df['Date']
```

```python
numerical_cols = df.select_dtypes(exclude='object').columns
numerical_cols
```

```python
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols
```

```python
exclude_cols = [
    'OrderID',
    'CustomerID',
    'TrackingNumber',
    'CouponCode'
]

text_clean_cols = [
    col for col in categorical_cols
    if col not in exclude_cols
]
    
print(text_clean_cols)
    
for col in text_clean_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )
```

```python
df['TrackingNumber']
```

```python
df.head()
```

```python
df['CouponCode'] = df['CouponCode'].fillna(
    'No Coupon'
)
```

```python
df['CouponCode']
```

```python
df['ReferralSource'] = df['ReferralSource'].fillna(
    'Unknown'
)
```

```python
df.head()
```

```python
df[
    (df['OrderStatus']=='Delivered')
    &
    (df['TrackingNumber'].isnull())
]
```

```python
df['UnitPrice'] = df['UnitPrice'].fillna(
    df['UnitPrice'].median()
)
```

```python
df['Quantity'] = df['Quantity'].fillna(
    df['Quantity'].median()
)
```

```python
df[df['Quantity'] <= 0]
```

```python
df[df['UnitPrice'] <= 0]
```

```python
df.head()
```

```python
df['CalculatedTotalPrice'] = (
    df['Quantity'] * df['UnitPrice']
)
```

```python
mismatch_rows = df[
    abs(df['CalculatedTotalPrice'] - df['TotalPrice']) > 0.01
]

mismatch_rows
```

```python
df['PriceValidation'] = (
    abs(
        (df['Quantity'] * df['UnitPrice'])
        - df['TotalPrice']
    ) <= 0.01
)
```

```python
df['PriceValidation'].value_counts()
```

```python
invalid_orders = df[
    df['PriceValidation'] == False
]

invalid_orders
```

```python
df['PriceDifference'] = (
    df['TotalPrice']
    -
    (df['Quantity'] * df['UnitPrice'])
)
```
